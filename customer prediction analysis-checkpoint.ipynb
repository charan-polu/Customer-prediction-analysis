Load and Preprocess the data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Define the data (one record for demonstration purposes)
df = pd.read_csv("C:\\Users\\HARSHITH ATIKETI\\Downloads\\customer_purchase_data.csv")
print(df)


X = df.drop('PurchaseStatus', axis=1)
y = df['PurchaseStatus']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
#importing Libraries
import ipywidgets as widgets

from IPython.display import display
```

```python
#Creating Widgets
viz_dropdown = widgets.Dropdown(
    options=[
        ('Distribution of Age', 'age_dist'),
        ('Count of Gender', 'gender_count'),
        ('Annual Income vs Number of Purchases', 'income_purchases'),
        ('Product Category vs Number of Purchases', 'category_purchases')
    ],
    value='category_purchases',   #to give default options to display the viz
    
    description='Select Visualization:'
)
```

```python
#Defining Visualizations
def plot_age_distribution():
    plt.figure(figsize=(8, 6))
    sns.histplot(df['Age'], kde=True, bins=20)
    plt.title('Distribution of Age')
    plt.show()

def plot_gender_count():
    plt.figure(figsize=(8, 6))
    sns.countplot(x='Gender', data=df)
    plt.title('Count of Gender')
    plt.show()

def plot_income_vs_purchases():
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x='AnnualIncome', y='NumberOfPurchases', hue='PurchaseStatus', data=df)
    plt.title('Annual Income vs Number of Purchases')
    plt.show()

def plot_category_vs_purchases():
    plt.figure(figsize=(10, 6))
    sns.barplot(x='ProductCategory', y='NumberOfPurchases', hue='PurchaseStatus', data=df)
    plt.title('Product Category vs Number of Purchases')
    plt.show()
```

```python
#Creating a Function to Display the Selected Visualization
def update_viz(selected_viz):
    if selected_viz == 'age_dist':
        plot_age_distribution()
    elif selected_viz == 'gender_count':
        plot_gender_count()
    elif selected_viz == 'income_purchases':
        plot_income_vs_purchases()
    elif selected_viz == 'category_purchases':
        plot_category_vs_purchases()

widgets.interactive(update_viz, selected_viz=viz_dropdown)
```

```python
display(viz_dropdown)
update_viz(viz_dropdown.value)
```

```python
# Train logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
AccuracyScore = accuracy_score(y_test,y_pred)
print(AccuracyScore)
```

```python
# Train random forest model
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
# Predict on test set
y_pred_rf = rf.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)

print(accuracy_rf)
```

```python
# Predicting with new data using logistic regression
newdata = [[40,1,66120,5,0,30.5,0,5]]
sol = model.predict(newdata)
if sol==0:
    print("Customer will Buy")
else:
    print("Customer may or may not buy")
```

```python
# Predicting with new data using random forest
newdata = [[11,1,67000,2,4,21,0,0]]
sol = rf.predict(newdata)
if sol==0:
    print("Customer will Buy")
else:
    print("Customer may or may not buy")
```
